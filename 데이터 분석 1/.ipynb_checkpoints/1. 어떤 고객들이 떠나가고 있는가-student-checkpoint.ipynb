## 1. Loading data

#### 1) 필요한 라이브러리들을 호출

```python
#데이터 분석에 필요한 패키지들을 파이썬으로 호출
import pandas as pd#pandas 패키지 호출 & 그이름은= pd
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

#### 2) 각 변수별로 데이터 할당 

```python
# 데이터 불러오기
dau = pd.read_csv("dau.csv")#함수 라이브러리 이름.함수 & 함수 위에 커서 두고 좌 shift+tab은 함수 내용 보기
user = pd.read_csv("user_info.csv")#이 read 함수는 뒤에 자료형을 _~로 받는다(자료 읽어옴)
```

# 2. Data cleaning

#### 1) 각 데이터들을 결합

```python
# Dau 데이터에 user 데이터를 결합하기
DauUser = dau.merge(user,on=["user_id","app_name"]) #X.merge(Y,on=[?])는 X에 Y의 데이터를 ?를 공통 기준으로 해서 합친다.
```

```python
DauUser.head()
```

#### 2)월별 정렬을 위한 문자 처리

```python
DauUser["log_month"]=DauUser["log_date"].str[0:7]
#변수["?"]는 ?슬롯 추가 의미(기존에 ?슬롯이 없을 시) or 기존에 ? 슬롯이 있으면 ?슬롯만 보이기
#str[c:b]는 c~b까지 나타내기
```

# 3. Data analysis

#### 1)크로스 집계를 이용한 분석

#### 성별에 따른 접속 일자 

```python
pd.crosstab(DauUser["log_month"],DauUser["gender"]) #crosstab은 ()의기준으로 분류
```

#### 연령에 따른 접속일자

```python
pd.crosstab(DauUser["log_month"],DauUser["generation"])
```

20대 6000,30대 6000명 이탈

#### 성별-연령에 따른 접속일자 - 첫 번째 방법

```python
pd.crosstab(DauUser["log_month"],[DauUser["gender"],DauUser["generation"]])
```

#### 디바이스 유형에 따른 접속일자

```python
pd.crosstab(DauUser["log_month"],DauUser["device_type"])
```

# 4. Data Visualization

#### 1) 디바이스 유형별 접속일자의 시각화 - 라인 그래프

```python
data=pd.crosstab(DauUser["log_date"],DauUser["device_type"])
```

```python
graph=pd.DataFrame(data)
graph.plot.line(figsize=(15,5))
```

#### 2) 디바이스 유형별 접속일자의 시각화 - Heat Map

```python
sns.set(rc={"figure.figsize":(10,20)})
sns.heatmap(data)
```
